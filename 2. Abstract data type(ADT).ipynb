As the name says abstract, it only tells basic info like type of data stored, the operations supported on them. An ADT specifies what each operation does, but not how it does it.

this means, it will tell only when we raise the excelerator vichele moves faster but it will not tell how it's happening.

one more example is, i am travelling to hyderabad from tenali. this is like abstract data type because it's only telling that i am going to travel to some place but it's not telling how i am going to travel. there are many ways of travelling by walk, running, bike, car, bus, train and flight.
